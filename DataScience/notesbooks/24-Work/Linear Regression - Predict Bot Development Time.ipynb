```python
import pandas as pd
import numpy as np
```

```python
botData = pd.read_csv("CompletedBots.csv")
```

```python
botData.head()
```

```python
print(botData.columns)
botData = botData.dropna()
botData.loc[:, 'Annual Financial Benefits (Estimated)':'Development Time (Business Days)']
```

```python
X = botData[['Rules Based', 'Stability', 'Data', 'Complexity',
       'Applications', 'Errors', 'Risk', 'Save Time', 'Frequency',
       'Save Money', 'Avoid Spending', 'Avoid Risks', 'Customers', 'Employees',
       'Operations (LOB)', 'Other', 'Ease of Implementation',
       'Potential Benefits', 'Development Effort (Weeks)',
       'One Time Costs (Estimated)', 'Pay Back Period','Potential Hours Saved', 'Development Effort (Estimated)','Financial Quantification(Efficiency Gains)','Forecast Impact -Total - First Year (Estimated)',
       'ROA']]

y = botData['Development Time (Business Days)']

X = X.apply(pd.to_numeric, errors='coerce')
y = y.apply(pd.to_numeric, errors='coerce')
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
# The coefficients
print('Coefficients: \n', lm.coef_)
```

```python
predictions = lm.predict( X_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
sns.distplot((y_test-predictions),bins=50);
```
